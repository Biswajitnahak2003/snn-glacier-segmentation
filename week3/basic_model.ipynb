```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import torch
```

```python
df = pd.read_csv(r"D:\downloads\archive\breast-cancer.csv")
df.head()
```

```python
df = df.drop(columns=['id'])
df.head()
```

```python
df.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.drop(columns=['diagnosis']), df['diagnosis'], test_size=0.2, random_state=42)
```

```python
scalar = StandardScaler()
x_train = scalar.fit_transform(x_train)
x_test = scalar.transform(x_test)
```

```python
print(x_train)
print(y_train)
```

```python
le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)
```

```python
print(y_train)
```

```python
x_train_tensor, x_test_tensor, y_train_tensor, y_test_tensor = torch.from_numpy(x_train),torch.from_numpy(x_test), torch.from_numpy(y_train), torch.from_numpy(y_test)
x_train_tensor, x_test_tensor, y_train_tensor, y_test_tensor = x_train_tensor.float(), x_test_tensor.float(), y_train_tensor.float(), y_test_tensor.float()
```

```python
x_train_tensor
```

```python
import torch.nn as nn
class MyNeuralNetwork(nn.Module):
    def __init__(self, num_features):
        # self.weights = torch.rand(x.shape[1], 1, dtype = torch.float64, requires_grad=True)
        # self.bias = torch.zeros(1, dtype = torch.float64, requires_grad=True)
        super().__init__()
        self.network = nn.Sequential(
            torch.nn.Linear(num_features, 1),
            torch.nn.Sigmoid()
        )
    def forward(self, features):
       out = self.network(features)
       return out
    # def loss(self, y_pred, y):
    #     epsilon = 1e-7
    #     y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)
    #     loss = -torch.mean(y * torch.log(y_pred) + (1 - y) * torch.log(1 - y_pred))
    #     return loss
```

```python
lr = 0.01
epochs = 25
```

```python
loss_function = nn.BCELoss()
```

```python
# training
model = MyNeuralNetwork(x_train.shape[1])

# optimizer 
optimizer = torch.optim.SGD(model.parameters(), lr=lr)

## loop 
for epoch in range (epochs):

    ## forward pass
    y_pred = model(x_train_tensor)
    ## loss
    loss = loss_function(y_pred, y_train_tensor.view(-1,1))

    ## zero grad
    optimizer.zero_grad()

    ## backward pass
    loss.backward()
    
    ## update weights and bias
    # with torch.no_grad():
    #     model.weights -= lr * model.weights.grad
    #     model.bias -= lr * model.bias.grad
    optimizer.step()
    ## zero grad    
    # model.weights.grad.zero_()
    # model.bias.grad.zero_()
     ## print loss
    print(f"epoch {epoch+1}, loss = {loss.item()}")
```

```python
# evaluation
with torch.no_grad():
    y_pred = model.forward(x_test_tensor)
    y_pred = (y_pred > 0.5).float()
    accuracy = (y_pred.squeeze() == y_test).float().mean()
    print(f"accuracy = {accuracy.item()}")
```
